```python
import numpy as np
```

```python
class Logistic_regression:

    def __init__(self, learning_rate=0.01, no_of_iterations=1000):
        self.learning_rate = learning_rate
        self.no_of_iterations = no_of_iterations

    def fit(self, x, y):
        # m = number of samples, n = number of features
        self.m, self.n = x.shape

        self.w = np.zeros(self.n)
        self.b = 0
        self.x = x
        self.y = y

        # Gradient Descent
        for i in range(self.no_of_iterations):
            self.update_weights()

    def update_weights(self):
        # Predictions using sigmoid
        y_hat = 1 / (1 + np.exp(-(self.x.dot(self.w) + self.b)))  # shape (m,)

        # Gradients
        dw = (1 / self.m) * np.dot(self.x.T, (y_hat - self.y))  # shape (n,)
        db = (1 / self.m) * np.sum(y_hat - self.y)

        # Update parameters
        self.w -= self.learning_rate * dw
        self.b -= self.learning_rate * db

    def predict(self, x):
        # Predictions for new data
        y_pred = 1 / (1 + np.exp(-(x.dot(self.w) + self.b)))
        y_pred = np.where(y_pred > 0.5, 1, 0)
        return y_pred
```

```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
scaler=StandardScaler()
```

```python
x_scale=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scale,y,test_size=0.2,stratify=y,random_state=2)
```

```python
model = Logistic_regression(learning_rate=0.01, no_of_iterations=1000)
model.fit(x_train, y_train)
predictions = model.predict(x_test)
```

```python
acc=accuracy_score(y_test,predictions)
print(acc)
```
