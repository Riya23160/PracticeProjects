```python
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
diabetes=pd.read_csv('diabetes.csv')
```

```python
diabetes.head()
```

```python
diabetes.shape
```

```python
diabetes.groupby('Outcome').mean()
```

```python
diabetes.value_counts('Outcome')
```

```python
y=diabetes['Outcome']
x=diabetes.drop(columns='Outcome',axis=1)
print(x)
print(y)
```

Standardising

```python
scaler=StandardScaler()
```

```python

scaler.fit(x)
standardized_data=scaler.transform(x)
print(standardized_data)
```

```python
x=standardized_data
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train,y_train)
```

```python
x_train_pred=classifier.predict(x_train)
accur=accuracy_score(x_train_pred,y_train)
print(accur)
```

```python
 x_test_pred=classifier.predict(x_test)
 test_data_acc=accuracy_score(x_test_pred,y_test)
 print(test_data_acc)
```

```python
inp=[4,110,92,0,0,37.6,0.191,30]
inp_as_np=np.asarray(inp)
input=inp_as_np.reshape(1,-1)
std=scaler.transform(input)

print(std)

prediction=classifier.predict(std)
print(prediction)

if(prediction[0]==0):
  print("person is not diabetic")

else:
  print("person is diabetic")
```
