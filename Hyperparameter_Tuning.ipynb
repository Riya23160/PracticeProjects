```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
dfa=sklearn.datasets.load_breast_cancer()
```

```python
print(dfa)
```

```python
df=pd.DataFrame(dfa.data,columns=dfa.feature_names)
```

```python
df.head()
```

```python
df['label']=dfa.target
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df['label'].value_counts()
```

```python
x=df.drop('label',axis=1)
y=df['label']
```

```python
x=np.asarray(x)
y=np.asarray(y)
```

#gridsearch cv is used for determining trhe best parameters by searching in a grid like manner


```python
model=SVC()
```

```python
parameters={
    'kernel':['linear','rbf','poly','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
classifier=GridSearchCV(model,parameters,cv=5)
```

```python
#fitting data in model
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
best_para=classifier.best_params_
print(best_para)
```

```python
highest_accuracy=classifier.best_score_
print(highest_accuracy)
```

```python
#loadin result to panda dataframe
result=pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_resultt=result[['param_C','param_kernel','mean_test_score']]
print(grid_search_resultt)
```

highest accuracy=95.2
and best value= {c:10,kernel:linear}

#Randomized Search CV

```python
model=SVC()
```

```python
para={
    'kernel':['linear','rbf','poly','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
classifier=RandomizedSearchCV(model,para,cv=5)
```

```python
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
best_para=classifier.best_params_
print(best_para)
```

```python
highest_accuracy=classifier.best_score_
print(highest_accuracy)
```

```python
result=pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
randomized_search_resultt=result[['param_C','param_kernel','mean_test_score']]
print(randomized_search_resultt)
```

highest accuracy= 95.0

para:(kernel: linear
c:10}
