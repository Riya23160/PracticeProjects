Dividing the dataframe into two different parts based on the class

```python
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
print(legit.shape,fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

UnderSampling

```python
legit_sample=legit.sample(n=492)
```

```python
ds=pd.concat([legit_sample,fraud],axis=0)
```

```python
ds.head()
```

```python
ds.value_counts('Class')
```

```python
ds.groupby('Class').mean()
```

```python
ds.shape
```

```python
x=ds.drop(columns='Class',axis=1)
y=ds['Class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
train_pred=model.predict(x_train)
acc=accuracy_score(train_pred,y_train)
print(acc)
```

```python
test_pred=model.predict(x_test)
accscore=accuracy_score(test_pred,y_test)
print(accscore)
```

```python
input=x_test.iloc[[3]]
print(y_test.iloc[3])
prediction=model.predict(input)
print(prediction)
if(prediction[0]==0):
  print('Fraud')
else:
  print('Legitimate')
```
